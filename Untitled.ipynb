```python
import random
import math
import numpy as np
import matplotlib.pyplot as plt
```

Set up arrivals

```python
def compute_arrival_times(journal_rate, n_journal_entries):
    # exponential distribution of inter-arrival delays
    inter_arrival_delays = -np.log(1.0 - np.random.rand(n_journal_entries)) / journal_rate
    return np.cumsum(inter_arrival_delays)

def assign_journal_entries_to_books(n_books, n_journal_entries):
    return np.random.randint(0, n_books, n_journal_entries)
```

```python
plt.plot(compute_arrival_times(0.4, 100))
```

```python
def simulate(n_journal_entries, n_books, journal_rate, per_journal_latency):
    arrival_times = compute_arrival_times(journal_rate, n_journal_entries)
    buckets = assign_journal_entries_to_books(n_books, n_journal_entries)
    waiting_times = []

    for book in range(0, n_books):
        per_book_arrival_times = arrival_times[buckets == book]
        n = len(per_book_arrival_times)
        processing_times = [0]
        for i in range(n):
            processing_times.append(max(per_book_arrival_times[i], processing_times[-1] + per_journal_latency))
        processing_times = processing_times[1:]
        waiting_times += list(processing_times - per_book_arrival_times)
        
    return np.array(waiting_times)
```

```python
# returns the (50, 75, 95, 99, 99.9) percentiles
def important_percentiles(n_books, journal_rate, per_journal_latency):
    n_journal_entries = int(journal_rate * per_journal_latency * 100)
    waiting_times = simulate(n_journal_entries, n_books, journal_rate, per_journal_latency)
    return np.percentile(waiting_times, [50, 75, 95, 99, 99.9])
```

```python
n_books_samples = np.geomspace(10, 1000, num=20, dtype='int32')
datapoints = 20
latencies_by_nbooks = np.zeros((datapoints, 5))
journal_latency = 1.0

for i, n_books in enumerate(n_books_samples):
    saturated_journal_rate = n_books / journal_latency
    latencies_by_nbooks[i] = important_percentiles(n_books, 0.5 * saturated_journal_rate, journal_latency)
    
plt.title("Journaling latencies")
plt.plot(n_books_samples, latencies_by_nbooks[:, 0])
plt.plot(n_books_samples, latencies_by_nbooks[:, 1])
plt.plot(n_books_samples, latencies_by_nbooks[:, 2])
plt.plot(n_books_samples, latencies_by_nbooks[:, 3])
plt.plot(n_books_samples, latencies_by_nbooks[:, 4])
plt.xlabel("2N Books handling N concurrent journal requests on avg")
plt.ylabel("Latency")
```
